```python
data_grp2 = data.iloc[:,3:73]
data_grp2
```

```python
data_grp2 = data_grp2.groupby( 'LIBREG').mean()
data_grp2
```

```python
data_grpT = data_grp2.transpose()
data_grpT
```

```python
data_grpT = data_grpT.reset_index()
data_grpT
```

```python
data_grpT = data_grpT.rename(columns={"index": "periode"})
data_grpT
```

```python
# Après avoir visualisé le contenu de data2, utiliser pandas.plot() pour tracer la ligne
# représentant l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte
# d'Azur

data_grpT.plot(x = 'periode' , y = "Provence-Alpes-Côte d'Azur", figsize = (15, 10))
```

```python
data_grpT.plot(x = 'periode' , figsize = (15, 10))
```

```python
# Représenter l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte
# d'Azur’, ‘Occitanie’ et ‘Ile-de-France’ sur un même graphique.

data_grpT.plot(x = 'periode' , y = ["Provence-Alpes-Côte d'Azur","Occitanie","Ile-de-France"], figsize = (15, 10))
```
